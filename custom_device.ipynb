```python
# CustomDevice is implemented in custom_device.py
from custom_device import CustomDevice
from pycap import PropertyTree
ptree=PropertyTree()
ptree.put_double('initial_voltage',1)
device=CustomDevice(ptree)
```

```python
from pycap import CyclicChargeDischarge
ptree=PropertyTree()
ptree.put_string('start_with','charge')
ptree.put_int   ('cycles',2)
ptree.put_double('time_step',0.01)
ptree.put_string('charge_mode','constant_current')
ptree.put_double('charge_current',0.25)
ptree.put_string('charge_stop_at_1','voltage_greater_than')
ptree.put_double('charge_voltage_limit',2.1)
ptree.put_string('discharge_mode','constant_current')
ptree.put_double('discharge_current',0.5)
ptree.put_string('discharge_stop_at_1','voltage_less_than')
ptree.put_double('discharge_voltage_limit',0.7)
ccd=CyclicChargeDischarge(ptree)
```

```python
%matplotlib inline
from pycap import initialize_data,plot_data
data=initialize_data()
ccd.run(device,data)
plot_data(data)
```
